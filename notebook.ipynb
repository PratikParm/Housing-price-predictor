# Data Preprocessing

```python
# Changing data  types of certain columns
data['number of bathrooms'] = data['number of bathrooms'].astype(int)
data['number of floors'] = data['number of floors'].astype(int)
```

```python
# Dropping unncessary columns
data = data.drop(['id', 'Date', 'Postal Code', 'Area of the house(excluding basement)', 'Area of the basement'], axis=1)
```

The features `living area`, `Area of the house(excluding basement)` and `Area of the basement` are inter dependent. So as to keep the features independent of each other, we remove the later two features.

# Create Train-Test sets

As the dataset is relatively small, we use stratefied sampling to avoid sampling bias.

```python
data['living area'].hist()
```

```python
data['living area'].sort_values(ascending=False).head(10)
```

```python
# Creating a category based on living area
data["living_area_cat"] = pd.cut(data["living area"],
                               bins=[0., 1000, 2000, 3000, 4000, np.inf],
                               labels=[1, 2, 3, 4, 5])
```

```python
data["living_area_cat"].value_counts()
```

```python
data["living_area_cat"].hist()
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(data, data["living_area_cat"]):
    strat_train_set = data.loc[train_index]
    strat_test_set = data.loc[test_index]
```

```python
strat_test_set["living_area_cat"].value_counts() / len(strat_test_set)
```

```python
data["living_area_cat"].value_counts() / len(data)
```

```python
for set_ in (strat_train_set, strat_test_set):
    set_.drop("living_area_cat", axis=1, inplace=True)
```

# Preparing data for Machine Learning

```python
data = strat_train_set.drop("Price", axis=1) # drop labels for training set
data_labels = strat_train_set["Price"].copy()
```

There are no empty entires in this perticular data, but we will include a step in the data preparation timeline to replace missing entries in the training data with the median value of that field.

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy="median")
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy="median")),
        ('std_scaler', StandardScaler()),
    ])

data_prepared = pipeline.fit_transform(data)
```

# Selecting and training the model

## Linear Regressor

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(data_prepared, data_labels)
```

```python
some_data = data.iloc[:5]
some_labels = data_labels.iloc[:5]
some_data_prepared = pipeline.transform(some_data)

print("Predictions:", lin_reg.predict(some_data_prepared))
```

```python
from sklearn.metrics import mean_squared_error

data_predictions = lin_reg.predict(data_prepared)
lin_mse = mean_squared_error(data_labels, data_predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

## Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor()
tree_reg.fit(data_prepared, data_labels)
```

```python
data_predictions = tree_reg.predict(data_prepared)
tree_mse = mean_squared_error(data_labels, data_predictions)
tree_rmse = np.sqrt(tree_mse)
tree_rmse
```

### Cross validation

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(tree_reg, data_prepared, data_labels,
                    scoring="neg_mean_squared_error", cv=10)

tree_rmse_scores = np.sqrt(-scores)
```

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())
    
display_scores(tree_rmse_scores)
```

```python
tree_rmse_scores.mean()/data_labels.median()
```

## Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
forest_reg = RandomForestRegressor()
forest_reg.fit(data_prepared, data_labels)
```

```python
scores = cross_val_score(forest_reg, data_prepared, data_labels,
                    scoring="neg_mean_squared_error", cv=10)

forest_rmse_scores = np.sqrt(-scores)
display_scores(forest_rmse_scores)
```

Random forest regressor performs better than the other models in terms of RMSE

## Grid Search

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV

param_grid = [
    {'n_estimators': [20, 30, 40], 'max_features': [5, 8, 11, 14, 17]},
    {'bootstrap': [False], 'n_estimators': [10, 30], 'max_features': [5, 8, 11]}
]

forest_reg = RandomForestRegressor()

grid_search = GridSearchCV(forest_reg, param_grid, cv=5, scoring = 'neg_mean_squared_error',
                          return_train_score=True)

grid_search.fit(data_prepared, data_labels)
```
